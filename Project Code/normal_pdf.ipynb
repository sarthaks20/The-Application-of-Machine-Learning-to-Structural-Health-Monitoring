```python
import numpy as np
import pandas as pd
from scipy.stats import norm
import matplotlib.pyplot as plt
%matplotlib widget
```

```python
state = np.zeros((5, 100001, 8))
state[0] = (pd.read_csv("datasets/state0_datasets/state0_acceleration_timehistory2.csv", index_col = "Time")).to_numpy()
state[1] = (pd.read_csv("datasets/state1_datasets/state1_acceleration_timehistory2.csv", index_col = "Time")).to_numpy()
state[2] = (pd.read_csv("datasets/state2_datasets/state2_acceleration_timehistory2.csv", index_col = "Time")).to_numpy()
state[3] = (pd.read_csv("datasets/state3_datasets/state3_acceleration_timehistory2.csv", index_col = "Time")).to_numpy()
state[4] = (pd.read_csv("datasets/state4_datasets/state4_acceleration_timehistory2.csv", index_col = "Time")).to_numpy()
```

```python
mean = np.zeros((5, 8))
for i in range(5):
    mean[i, :] = np.mean(state[i], axis = 0)
```

```python
std = np.zeros((5, 8))
for i in range(5):
    std[i, :] = np.std(state[i], axis = 0)
```

```python
fig = plt.figure(figsize=(14, 8), constrained_layout = True)
ax = fig.subplots(nrows = 2, ncols = 4, sharey = 'row')

for i in range(4):
    for j in range(5):
        arr = list(state[j][:, i])
        arr.sort()
        ax[0][i].plot(arr, norm.cdf(arr, mean[j][i], std[j][i]), label = f"State #{j}", linewidth = 0.6)
        ax[0][i].set_xlabel("Acceleration")
        #ax[0][i].set_xlim([-1.0,1.0])
        ax[0][i].set_ylabel("CDF")
        ax[0][i].set_title(f"m{i+1}")
        ax[0][i].legend(loc="lower right")

for i in range(4,8):
    for j in range(5):
        arr = list(state[j][:, i])
        arr.sort()
        ax[1][i-4].plot(arr, norm.cdf(arr, mean[j][i], std[j][i]), label = f"State #{j}", linewidth = 0.6)
        ax[1][i-4].set_xlabel("Acceleration")
        #ax[0][i].set_xlim([-1.0,1.0])
        ax[1][i-4].set_ylabel("CDF")
        ax[1][i-4].set_title(f"m{i+1}")
        ax[1][i-4].legend(loc="lower right")
#ax[1][0].legend(loc='center', bbox_to_anchor=(0.5, -0.4), fancybox=True, shadow=True, ncol=4)
plt.savefig("cdf.svg")
'''
for i in range(8):
    for j in range(5):
        arr = list(state[j][:, i])
        arr.sort()
        ax[1][i].plot(arr, norm.cdf(arr, mean[j][i], std[j][i]), label = f"State #{j}", linewidth = 0.2)
        ax[1][i].set_xlabel("Acceleration")
        #ax[1][i].set_xlim([-1.0,1.0])
        ax[1][i].set_ylabel("CDF")
        #ax[1][i].legend(loc = "lower right")
#ax[1][4].legend(loc='upper center', bbox_to_anchor=(0.5, -0.4), fancybox=True, shadow=True, ncol=8)
'''
```
